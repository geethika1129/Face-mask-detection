```python
datai=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
datai.detectMultiScale(img)
```

```python
while True:
    faces=datai.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
    cv2.imshow('result',img)
    if cv2.waitKey(2)==27:
        break
cv2.destroyAllWindows()
```

```python
capture=cv2.VideoCapture(1)
while True:
    flag,img=capture.read()
    if flag:
        faces=datai.detectMultiScale(img)
        for x,y,w,h in faces:
           cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
    cv2.imshow('result',img)
    if cv2.waitKey(2)==27:
        break
cv2.destroyAllWindows()
            
```
